```python
import os
import pandas as pd
from google.cloud import bigquery

# 1. Point Python to secret key
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "credentials.json"

# 2. Set Project ID
PROJECT_ID = "olist-analytics-491319" # 
DATASET_ID = "raw_data"

# 3. Connect to BigQuery
client = bigquery.Client()

# 4. Get the path to the folder where CSVs are
folder_path = "./data" 

# 5. Loop through the folder and upload every CSV
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        # Clean up the file name to use as the table name
        table_name = file.replace("olist_", "").replace("_dataset.csv", "").replace(".csv", "")
        table_id = f"{PROJECT_ID}.{DATASET_ID}.{table_name}"
        
        print(f"Uploading {file} to BigQuery...")
        
        # Read the CSV with Pandas
        df = pd.read_csv(os.path.join(folder_path, file))
        
        # Upload to BigQuery
        job_config = bigquery.LoadJobConfig(write_disposition="WRITE_TRUNCATE")
        job = client.load_table_from_dataframe(df, table_id, job_config=job_config)
        job.result() # Wait for the job to finish
        
        print(f"Successfully loaded {len(df)} rows into {table_name}!\n")
```
